# Phase 1: Exploratory Data Analysis (EDA) and Preprocessing

This notebook covers the initial analysis of the credit card transaction dataset, handling class imbalance, and preprocessing the data for model training.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
import os

%matplotlib inline
```

## 1. Load Dataset

```python
data_path = '../data/creditcard.csv'
df = pd.read_csv(data_path)
df.head()
```

## 2. Exploratory Data Analysis

### 2.1 Class Distribution
The dataset is highly imbalanced, which is common in fraud detection. Most transactions are legitimate (Class 0), while only a small fraction are fraudulent (Class 1).

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Class', data=df)
plt.title('Class Distribution (0: Not Fraud, 1: Fraud)')
plt.show()

print("Class Distribution:")
print(df['Class'].value_counts(normalize=True) * 100)
```

### 2.2 Transaction Amount Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(df[df['Class'] == 1]['Amount'], bins=50, color='red', label='Fraud', kde=True)
sns.histplot(df[df['Class'] == 0]['Amount'], bins=50, color='blue', label='Not Fraud', kde=True)
plt.title('Transaction Amount Distribution')
plt.legend()
plt.show()
```

### 2.3 Correlation Heatmap

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), cmap='coolwarm', annot=False)
plt.title('Correlation Heatmap')
plt.show()
```

## 3. Preprocessing

### 3.1 Feature Scaling
The 'Amount' and 'Time' features have different scales compared to the PCA-transformed features (V1-V28). We'll scale 'Amount' and 'Time' using `StandardScaler`.

```python
scaler = StandardScaler()
df['Amount'] = scaler.fit_transform(df[['Amount']])
df['Time'] = scaler.fit_transform(df[['Time']])
```

### 3.2 Train-Test Split
We'll use a stratified split to maintain the class distribution in both sets.

```python
X = df.drop('Class', axis=1)
y = df['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

### 3.3 Handling Class Imbalance (SMOTE)
SMOTE (Synthetic Minority Over-sampling Technique) creates synthetic samples of the minority class to balance the dataset.

```python
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

print("Before SMOTE:", y_train.value_counts())
print("After SMOTE:", y_train_res.value_counts())
```
